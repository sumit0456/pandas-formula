```python
df2 = pd.read_csv("D:\\data science\\data set\\weatherr.csv")
df2
type(df2.day[0])
```

```python
# since column 'day' is a string so we will change into date format and for this we use
df2 = pd.read_csv("D:\\data science\\data set\\weatherr.csv" , parse_dates = ['day'])
df2.head(3)
```

```python
# Here we will set day column to index column for this we use the command:
df2.set_index('day', inplace=True)
df2
```

```python
new_df2 = df2.fillna(0)
new_df2.head()
```

```python
new_df2 = df2.fillna({
          'temperature' : 0,
          'windspeed' : 0,
          'event' : 'No Events'
})
new_df2.head()
```

```python
# ffill() is used to fill the null value as just previous value: lets see the example
new_df2 = df2.fillna(method='ffill')
new_df2.head()
```

```python
# In the above code we can see that when we use ffill(),it is filling all next null value with the same value of previous one.
# But if we want to fill only 1 null value with the previous one then we use:--
new_df2 = df2.fillna(method='ffill',limit=1)
new_df2
```

```python
# bfill() is used to fill the null value as just next value: lets see the example
new_df2 = df2.fillna(method='bfill')
new_df2.head()
```

```python
new_df2 = df2.interpolate()
new_df2
```

```python
# it is more accurate interpolate(method='time') rather than interpolate(),bcoz it check near by the time or date and give the value
new_df2 = df2.interpolate(method='time')
new_df2
```

```python
# dropna() will remove all rows and columns in which there is atleast single NaN value.
new_df2 = df2.dropna()
new_df2
```

```python
# by this argument how='all' will remove those columns only in which all columns are NaN values.
new_df2 = df2.dropna(how='all')
new_df2
```

# Basic Concatenation

```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
# Ignore Index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
# concatenation and keys
df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
df.loc["us"]
```

```python
# Concatenation using index
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

```python
#Concatenate dataframe with series
s = pd.Series(["Humid","Dry","Rain"], name="event")
s
```

```python
df = pd.concat([temperature_df,s],axis=1)
df
```
